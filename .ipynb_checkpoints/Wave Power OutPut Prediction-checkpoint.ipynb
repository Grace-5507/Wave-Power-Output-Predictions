```python
def reshape_wec_data(df, label=None):
    """
    Convert wide-format WEC dataset into long-format.
    Keeps per-WEC (X, Y, Power) and farm-level (Total_Power, qW).
    """
    positions = []
    n = len([col for col in df.columns if col.startswith("X")])  # number of WECs
    
    for i, row in df.iterrows():
        total_power = row["Total_Power"]  # farm-level total power
        q_val = row["qW"]                # farm-level q-factor
        for j in range(n):
            x = row[f"X{j+1}"]
            y = row[f"Y{j+1}"]
            p = row[f"Power{j+1}"]
            # include farm-level columns too
            positions.append([x, y, p, i, label, total_power, q_val])
    
    pos_df = pd.DataFrame(
        positions, 
        columns=["X", "Y", "Power", "Config", "Farm", "Total_Power", "qW"]
    )
    return pos_df

```

```python
reshape_wec_data(df_SD100, label="100-WEC")
```

```python
#Evaluate all models



def evaluate_model(name, y_test, y_pred):
    print(f"\n{name}")
    print("R²:", r2_score(y_test, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
    print("MAE:", mean_absolute_error(y_test, y_pred))

evaluate_model("Decision Tree", y_test, y_pred)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("Gradient Boosting", y_test, y_pred_gb)
evaluate_model("Linear Regression", y_test, y_pred_lr)
```

```python
#Best performers so far → Decision Tree & Random Forest

#Random Forest Tuning with RandomizedSearchCV

# Define parameter grid
param_dist = {
    "n_estimators": [100, 200, 500],          # number of trees
    "max_depth": [10, 20, 30, None],         # depth of each tree
    "min_samples_split": [2, 5, 10],         # minimum samples to split a node
    "min_samples_leaf": [1, 2, 4],           # minimum samples per leaf
    "max_features": ["auto", "sqrt"]         # number of features to consider at each split
}
```

```python
rf = RandomForestRegressor(random_state=42)

# Randomized search
rf_random = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=10,              # number of random combos to try
    cv=3,                   # 3-fold cross-validation
    verbose=2,
    random_state=42,
    n_jobs=-1
)

rf_random.fit(X_train, y_train)
```

```python
print("Best parameters:", rf_random.best_params_)
```

```python
best_rf = rf_random.best_estimator_

y_pred = best_rf.predict(X_test)

print("R²:", r2_score(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

```python
NOTE:   The tuning didn’t change results much → this usually means the original RF was already close to optimal.

        Random Forests are very robust; even default settings often perform really well.

        Small gains (RMSE ↓ 5, MAE ↓ 2) are not huge, but they confirm your model is stable.
```
