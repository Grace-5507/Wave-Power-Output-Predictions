```python
def reshape_wec_data(df, label=None):
    """
    Convert wide-format WEC dataset into long-format.
    Keeps per-WEC (X, Y, Power) and farm-level (Total_Power, qW).
    """
    positions = []
    n = len([col for col in df.columns if col.startswith("X")])  # number of WECs
    
    for i, row in df.iterrows():
        total_power = row["Total_Power"]  # farm-level total power
        q_val = row["qW"]                # farm-level q-factor
        for j in range(n):
            x = row[f"X{j+1}"]
            y = row[f"Y{j+1}"]
            p = row[f"Power{j+1}"]
            # include farm-level columns too
            positions.append([x, y, p, i, label, total_power, q_val])
    
    pos_df = pd.DataFrame(
        positions, 
        columns=["X", "Y", "Power", "Config", "Farm", "Total_Power", "qW"]
    )
    return pos_df

```

```python
reshape_wec_data(df_SD100, label="100-WEC")
```

```python
pos_49 = reshape_wec_data(df_SD49, label="49-WEC")
print(pos_49.head())
```

```python
#Descision Tree

model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
#Predict & evaluate
from sklearn.metrics import r2_score, mean_squared_error
y_pred = model.predict(X_test)

print("R²:", r2_score(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))
```

```python
#Evaluate all models



def evaluate_model(name, y_test, y_pred):
    print(f"\n{name}")
    print("R²:", r2_score(y_test, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
    print("MAE:", mean_absolute_error(y_test, y_pred))

evaluate_model("Decision Tree", y_test, y_pred)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("Gradient Boosting", y_test, y_pred_gb)
evaluate_model("Linear Regression", y_test, y_pred_lr)
```
