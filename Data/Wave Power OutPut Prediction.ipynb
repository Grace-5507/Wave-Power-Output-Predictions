```python
position = []

for i, row in df_SD49.iterrows():
    n = len([col for col in df_SD49.columns if col.startswith("X")])
    for j in range(n):
        x=row[f"X{j+1}"]
        y=row[f"Y{j+1}"]
        p=row[f"Power{j+1}"]
        position.append([x,y,p])

pos_df = pd.DataFrame(position, columns=["X", "Y", "Power"])

pos_df
```

```python
# Scatter plot: position vs power
plt.figure(figsize=(8,6))
plt.scatter(pos_df["X"], pos_df["Y"], c=pos_df["Power"], cmap="viridis", alpha=0.7)
plt.colorbar(label="WEC Power Output")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("Individual WEC Positions vs Power Output")
plt.show()
```

```python

heatmap, xedges, yedges = np.histogram2d(pos_df["X"], pos_df["Y"],
                                         bins=50, weights=pos_df["Power"])
counts, _, _ = np.histogram2d(pos_df["X"], pos_df["Y"], bins=50)
avg_power = heatmap / np.maximum(counts, 1)

plt.imshow(avg_power.T, origin='lower', cmap="inferno",
           extent=[pos_df["X"].min(), pos_df["X"].max(),
                   pos_df["Y"].min(), pos_df["Y"].max()])
plt.colorbar(label="Average WEC Power")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("High-Power WEC Position Heatmap")
plt.show()
```
